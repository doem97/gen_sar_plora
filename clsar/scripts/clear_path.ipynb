```python
import os
import re
import shutil

def delete_files_with_threshold(folder_path, threshold):
    epoch_pattern = re.compile(r'epoch_(\d+)\.pth')

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            match = epoch_pattern.match(file)
            if match:
                epoch_index = int(match.group(1))
                if epoch_index < threshold:
                    file_path = os.path.join(root, file)
                    # print(file_path)
                    try:
                        os.remove(file_path)
                        print(f"Deleted file: {file_path}")
                    except Exception as e:
                        print(f"Error deleting file {file_path}: {e}")
```

```python
base = "/storage/tianzichen/DSO/mmclassification/outputs/"
folder_path = "res50_opensar"
folder_path = os.path.join(base, folder_path)
threshold = 80  # Replace this with your desired threshold value

delete_files_with_threshold(folder_path, threshold)
```
